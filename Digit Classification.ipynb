```python
# Imports

from sklearn.linear_model import Perceptron
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# Load the digits dataset and look at a few examples

digits = load_digits()
plt.gray()

for i in range(10):
    print('Digit = %d' % i)
    plt.matshow(digits.images[i])
    plt.show()
```

```python
# Get the attribute vectors and the class labels

X = digits.data
y = digits.target
```

```python
# Convert 10 digit problem to one in which the goal is to tell if a digit is
# 0 or not 0

y = [1 if a == 0 else 0 for a in y]
```

```python
# Split the data into a training set and a test set

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.5,
                                                    random_state=42)
```

### Perceptron implementation from scikit

```python
# If you name your classifier object clf as below then the code for the next task
# not need to be modified.
clf = Perceptron()

#train the perceptron on the training data
clf.fit(X_train, y_train)

#calculate training set accuracy
train_accuracy = clf.score(X_train, y_train)
print("Training Set Accuracy:", train_accuracy)

#calculate testing set accuracy
test_accuracy = clf.score(X_test, y_test)
print("Testing Set Accuracy:", test_accuracy)
```

# Result Comparison

The training set accuracy of 1.0 or 100% implies that the perceptron has successfully learned the training data and can perfectly classify all examples in the training set. On the other hand, the testing set accuracy of 0.9933.. or 99.33% suggests that the perceptron generalizes well but may encounter some difficulty with possible unseen data.
The high accuracy on both sets may indicate that the perceptron is really good in terms of the problem of distinguishing between digit 0 and non-digit 0. However, the slightly lower testing accuracy compared to the training accuracy could be attributed to the possibility of the model encountering more challenging examples or outliers that were not present in the training set.  Overall, the high accuracies indicate that the Perceptron is effective in solving classification of 0 and Non zero digits.


# Understanding Learned Weights in Zero Digit Classification

The learned weights make sense for the task of classifying zero digits as the positive class and all other digits as the negative class because its helps the perceptron to adjusts its weights to emphasize features that are important for distinguishing between the two classes through it's training process. For example, the weights for the pixels that make up the outer ring of the pixel image are largely positive. This means that when these pixels are activated they contribute positively to the output by making it more likely that the image will be classified as a "0". On the other hand, the weights for the pixels in the center of the pixel image are largely negative. This means that when these pixels are activated, they contribute negatively to the output, making it less likely that the image will be classified as a "0".  Overall, the learned weights effectively capture the key features of the "0" digit and distinguish it from other digits which makes it possible for the classifier to accurately identify "0"s in images. Therefore, the learned weights make sense as they highlight the discriminative features necessary for the perceptron to effectively classify zero digits as the positive class and all other digits as the negative class.
